```python
import pandas as pd

# import data from Excel
data = pd.read_excel("C:\\Users\\Tim\\Desktop\\Data Science\\Scriptie\\Tetris data\\code\\output_preprocessed_data.xlsx")

# print the data
print(data.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot histogram of 'average_score'
plt.hist(data['average_score'], bins=10, color='blue')

# Add labels and title
plt.xlabel('Final Tetris-Score')
plt.ylabel('Frequency')
plt.title('Histogram of Final Tetris-Score')

# Show the plot
plt.show()
```

```python
import scipy.stats as stats

# Calculate Spearman (data is non-normally distributed) correlation coefficients
spearman_corr, spearman_pval = stats.spearmanr(data['total_keystrokes_averagepergame'], data['average_score'])

# Create scatter plot
plt.scatter(data['total_keystrokes_averagepergame'], data['average_score'])
plt.xlabel('Total Number of Keystrokes')
plt.ylabel('Average Tetris-Score')
plt.title('Association between Total Number of Keystrokes and Average Tetris-Score')

# Add Pearson and Spearman correlation coefficients to the plot
plt.text(data['total_keystrokes_averagepergame'].min(), data['average_score'].max(), f"Spearman Correlation: {spearman_corr:.2f} (p-value: {spearman_pval:.2f})", ha='left', va='top', fontsize=10)

# Show the plot
plt.show()
```

```python
# Define old and new names
old_names = ['total_down_averagepergame', 'total_left_averagepergame', 'total_right_averagepergame', 'total_left_right_averagepergame', 'total_clockwise_rotations_averagepergame', 'total_conterclockwise_rotations_averagepergame', 'total_rotations_averagepergame']
new_names = ['Down keystroke', 'Left keystroke', 'Right keystroke', 'Left & right keystrokes', 'Clockwise rotation keystroke', 'Counterclockwise rotation keystroke', 'Total rotation keystrokes']

# Create a dictionary of old and new names
name_dict = dict(zip(old_names, new_names))

# Rename the columns
data_subset = data.rename(columns=name_dict)

# Select only the renamed columns
data_subset = data_subset[new_names]

# Calculate correlation matrix
corr_matrix = data_subset.corr(method='spearman')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", center=0, linewidths=0.5, vmin=0, vmax=1)

# Set title and labels
plt.title('Spearman Correlation Heatmap of the Keystrokes')
plt.xlabel('')
plt.ylabel('')

# Remove grid
plt.grid(False)

plt.show()
```

```python
# Renaming columns
data = data.rename(columns={
    'total_down_averagepergame': 'Down keystroke',
    'total_left_averagepergame': 'Left keystroke',
    'total_right_averagepergame': 'Right keystroke',
    'total_left_right_averagepergame': 'Left & right keystrokes',
    'total_clockwise_rotations_averagepergame': 'Clockwise rotation keystroke',
    'total_conterclockwise_rotations_averagepergame': 'Counterclockwise rotation keystroke',
    'total_rotations_averagepergame': 'Total rotation keystrokes',
    'average_score': 'Final Tetris-score'
})

# Calculate correlation matrix
data_subset = data[['Down keystroke', 'Left keystroke', 'Right keystroke', 'Left & right keystrokes', 'Clockwise rotation keystroke', 'Counterclockwise rotation keystroke', 'Total rotation keystrokes', 'Final Tetris-score']]
corr_matrix = data_subset.corr(method='spearman')

# Extract the correlations between the features and 'Final Tetris-score'
heatmap_data = corr_matrix[['Final Tetris-score']].iloc[:-1].sort_values(by='Final Tetris-score', ascending=False)

# Plot heatmap
plt.figure(figsize=(4, 6))
sns.heatmap(heatmap_data, annot=True, cmap="coolwarm", center=0, linewidths=0.5, cbar_kws={'label': 'Correlation'}, vmin=0, vmax=1)

# Set title and labels
plt.title('Spearman correlation of keystrokes and Final Tetris score')
plt.ylabel('Features')

plt.show()

```
