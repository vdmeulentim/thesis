```python
import pandas as pd

# import data from Excel
data = pd.read_excel("C:\\Users\\Tim\\Desktop\\Data Science\\Scriptie\\Tetris data\\code\\output_preprocessed_data.xlsx")

# print the data
print(data.head())
```

```python
from scipy.stats import pointbiserialr
import matplotlib.pyplot as plt
import seaborn as sns

# Set the plot style
sns.set(style='whitegrid')

# Create a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='tetris_experience_binary', y='average_score', palette='viridis')

continuous_var = data['average_score']
binary_outcome = data['tetris_experience_binary']

# Calculate the Point-Biserial correlation coefficient
corr, p_value = pointbiserialr(continuous_var, binary_outcome)

# Add the correlation coefficient and P-value to the plot
corr_text = f'Correlation coefficient: {corr:.3f}\nP-value: {p_value:.3f}'
plt.text(0.02, 0.98, corr_text, fontsize=12, va='top', ha='left', transform=plt.gca().transAxes)

# Customize the plot
plt.title('Correlation Analysis Final Tetris-Score vs Self-Assessed Experience Level')
plt.xlabel('Self-Assessed Experience Level')
plt.ylabel('Final Tetris-score')

# Replace the x-tick labels
plt.xticks([0, 1], ['Low', 'High'])

# Show the plot
plt.show()
```
