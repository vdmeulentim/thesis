```python
import pandas as pd

# import data from Excel
data = pd.read_excel("C:\\Users\\Tim\\Desktop\\Data Science\\Scriptie\\Tetris data\\code\\output_preprocessed_data.xlsx")

# print the data
print(data.head())
```

```python
from scipy.stats import pointbiserialr
import matplotlib.pyplot as plt
import seaborn as sns

# Set the plot style
sns.set(style='whitegrid')

# Create a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='tetris_experience_binary', y='total_keystrokes_averagepergame', palette='viridis')

# Customize the plot
plt.title('Box plot of Total Keystrokes Average per Game vs Self-Assessed Tetris Experience Level')
plt.xlabel('Self-Assessed Tetris Experience Level')
plt.ylabel('Total Keystrokes Average per Game')

# Show the plot
plt.show()
continuous_var = data['total_keystrokes_averagepergame']
binary_outcome = data['tetris_experience_binary']

# Calculate the Point-Biserial correlation coefficient
corr, p_value = pointbiserialr(continuous_var, binary_outcome)

print(f'Correlation coefficient: {corr:.3f}')
print(f'P-value: {p_value:.3f}')
```

```python
import pandas as pd

# Select the relevant columns
data_subset = data[['total_down_averagepergame', 'total_left_averagepergame', 'total_right_averagepergame', 'total_left_right_averagepergame', 'total_clockwise_rotations_averagepergame', 'total_conterclockwise_rotations_averagepergame', 'total_rotations_averagepergame']]

# Rename the columns with the new names
data_subset.columns = ['Down keystroke', 'Left keystroke', 'Right keystroke', 'Left & right keystrokes', 'Clockwise rotation keystroke', 'Counterclockwise rotation keystroke', 'Total rotation keystrokes']

# Calculate the Point-Biserial Correlation between the features and 'tetris_experience_binary'
correlations = [pointbiserialr(data_subset[feature], data['tetris_experience_binary']).correlation for feature in data_subset.columns]

# Create a DataFrame to store the correlation values
heatmap_data = pd.DataFrame(correlations, columns=['Self-Assessed Experience Level'], index=data_subset.columns)

# Sort the DataFrame in descending order
heatmap_data.sort_values(by='Self-Assessed Experience Level', ascending=False, inplace=True)

# Plot heatmap
plt.figure(figsize=(4, 6))
sns.heatmap(heatmap_data, annot=True, cmap="coolwarm", center=0, linewidths=0.5, cbar_kws={'label': 'Correlation'}, vmin=0, vmax=1)

# Set title and labels
plt.title('Point-Biserial Correlation of Keystrokes and Self-Assessed Experience Level')
plt.ylabel('Features')

plt.show()


```
